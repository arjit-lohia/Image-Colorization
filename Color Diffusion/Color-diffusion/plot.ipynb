```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV
df = pd.read_csv("C:/Users/Arjit/OneDrive/Desktop/MTP/CODE/Color Diffusion/Color-diffusion/logs/training_logs/version_5/metrics.csv")
```

```python
df_early = df[df['step'] <= 1000].iloc[::2]
df_mid = df[(df['step'] > 1000) & (df['step'] <= 3000)].iloc[::6]
df_late = df[df['step'] > 3000].iloc[::20]

# Combine sampled data
df_sampled = pd.concat([df_early, df_mid, df_late])
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_sampled['step'], df_sampled['total loss'], label='Total Loss', s=20)
plt.xlabel('Step')
plt.ylabel('Total Loss')
plt.title('Training Loss vs Step')
plt.grid(True)
plt.legend()
```
